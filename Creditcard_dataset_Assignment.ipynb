### Divide the time and amount columns of both fraude and non_fraude into batchs 

```python
fraude_records["Time"]
```

```python
round(max(non_fraude_records["Time"] / 1800))+1
```

```python
round(max(fraude_records["Time"] / 1800))+1
```

```python
# in case of time

print('-------------------------------- Fraude people ---------------------------')
print(f'Display the maximum number of time column = {max(fraude_records["Time"])}')
print(f'Display the minimum number of time column = {min(fraude_records["Time"])}')

# Each batch in time will contains the range from (0-1800) [means 1/2 hour] 
# get the Percentage for each batch in the time column of fraude people
start_index = 0
end_index = 1800
fraude_time_batches = dict()
for i in range(1, round(max(fraude_records["Time"]) / 1800)+2, 1):
    counter = fraude_records['Time'].between(left=start_index, right=end_index).sum()
    # fraude_time_batches[f'{(start_index/60) / 60} H - {(end_index/60) / 60} H'] = counter
    fraude_time_batches[i] = round(counter/len(fraude_records), 5) * 100
    
    start_index += 1800
    end_index += 1800

print(fraude_time_batches)
    
print()

print('-------------------------------- Non_Fraude people ---------------------------')
print(f'Display the maximum number of time column = {max(non_fraude_records["Time"])}')
print(f'Display the minimum number of time column = {min(non_fraude_records["Time"])}')

# get the Percentage for each batch in the time column of non fraude people
start_index = 0
end_index = 1800
non_fraude_time_batches = dict()
for i in range(1, round(max(non_fraude_records["Time"]) / 1800)+2, 1):
    counter = non_fraude_records['Time'].between(left=start_index, right=end_index).sum()
    # non_fraude_time_batches[f'{(start_index/60) / 60} H - {(end_index/60) / 60} H'] = counter
    non_fraude_time_batches[i] = round(counter/len(non_fraude_records), 5) * 100
    
    start_index += 1800
    end_index += 1800

print(non_fraude_time_batches)
```

```python
fraude_time_batches_data_frame = pd.DataFrame({'Batch':list(fraude_time_batches.keys()), 'Percentage': list(fraude_time_batches.values())})
clean_fraude_time_batches_data_frame = fraude_time_batches_data_frame[fraude_time_batches_data_frame['Percentage'] != 0]
clean_fraude_time_batches_data_frame
```

```python
non_fraude_time_batches_data_frame = pd.DataFrame({'Batch':list(non_fraude_time_batches.keys()), 'Percentage': list(non_fraude_time_batches.values())})
clean_non_fraude_time_batches_data_frame = non_fraude_time_batches_data_frame[non_fraude_time_batches_data_frame['Percentage'] != 0]
clean_non_fraude_time_batches_data_frame
```

### Scaling both columns Time and Amount

```python
data_scaled = data_without_outliers.copy()
```

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
data_scaled['Time'] = std.fit_transform(data_scaled[['Time']]).flatten()
data_scaled['Amount'] = std.fit_transform(data_scaled[['Amount']]).flatten()
```

### Class Balancing

```python
y_axis = list(data_scaled['Class'].value_counts())
x_axis = list(data_scaled['Class'].unique().astype(str))
print(x_axis)
print(y_axis)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(x=x_axis, height=y_axis)
plt.title('Unbalanced Classes')
plt.show()
```

```python
X = data_scaled.iloc[:,:-1]
Y = data_scaled.iloc[:,-1]
```

```python
X.shape
```

```python
Y.shape
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
X_oversampling, Y_oversampling = sm.fit_resample(X, Y)
```

### Data spliting

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_oversampling, Y_oversampling, test_size=0.2, random_state=5, stratify=Y_oversampling)
```

### Model Building and training

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)
```

### Model Evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, ConfusionMatrixDisplay
```

```python
y_predict = logistic_model.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_predict)
cm
```

```python
ConfusionMatrixDisplay(cm).plot()
plt.plot()
```

```python
accuracy = accuracy_score(y_test, y_predict)
accuracy
```

```python
report = classification_report(y_test, y_predict)
print(report)
```
